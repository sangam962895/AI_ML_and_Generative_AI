## Messages

## Use a Text Prompt When

- You have a single, standalone request.
- You don't need conversation history.
- You want minimal code complexity.

# Message Promt

Message prompt ka matlab hai prompt ko messages ki list ke form me dena, jahan har message ka ek role hota hai—jaise system, user, aur assistant.

# Message prompt tab useful hota hai jab:
-    conversation history maintain karni ho,
-   alag-alag roles/instructions dene hon,
-   multi-turn chatbot banana ho,
-   context ko structured form me bhejna ho.

# Common message types:

- System message — AI ko overall instructions aur behavior batata hai.
- User message — user ka question ya request hota hai.
- Assistant message — AI ka previous response hota hai.
- Tool message — kisi tool ya function ka result hota hai.

# System message

- System message ek high-level instruction hota hai jo AI ko batata hai ki kaise behave karna hai, kis role me answer dena hai, aur kin rules ko follow karna hai.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are a AI Expert."),
    HumanMessage("Hello, can you explain RAG?"),
]
response= model.invoke(messages)
print(response.content)
```

```python
## example 2
system_msg= SystemMessage("You are a helpful assistant for resume writing.")
messages = [
    system_msg,
    HumanMessage("Hello, can you help me write a resume?"),
]
response= model.invoke(messages)
print(response.content)
```

```python
## detailed example

from langchain.messages import SystemMessage, HumanMessage, AIMessage

#create an AI message
ai_msg = AIMessage("I had be happy to help you with that queastion")

# add to conversation history
messages = [
    SystemMessage("You are a helpful assistant for resume writing."),
    HumanMessage("Hello, can you help me write a resume?"),
    ai_msg,
    HumanMessage("Yes, I can help you with that. What kind of job are you applying for?")
]
response= model.invoke(messages)
print(response.content)
```

# Tool Calls

```python
from langchain.messages import AIMessage
from langchain.messages import ToolMessage

ai_message = AIMessage(
    content="",
    tool_calls=[
        {
            "name": "get_weather",
            "args": {"location": "Delhi"},
            "id": "1234",
            "type": "tool_call"
        }
    ]
)
```

```python
weather_result = "Sunny, 25°C"
tool_message = ToolMessage(
    content = weather_result,
    tool_call_id = "1234",
)
```

```python
messages = [
    HumanMessage("What's the weather like in Delhi?"),
    ai_message,
    tool_message,
]
response= model.invoke(messages)
```
