# tools

moel can request to call that perform tasks such as featching data from a database, searching the web, or running code tools are paring of.

```python
import os 
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:llama-3.1-8b-instant")
response = model.invoke("Write a eassy in 20 words")
response
```

```python
#tool

from langchain.tools import tool

@tool
def get_weather(location:str)->str:
    """Get the weather at a location"""
    return f"It is sunny in {location}"
model_with_tools=model.bind_tools([get_weather])
```

```python
response=model_with_tools.invoke("What the weather in Banglore")
print(response)
```

```python
response.tool_calls
```

# Tool Exicution Loop

```python
# Step 1 Model generate tool calls
message = [{"role": "user", "content": "What the weather in Bangalore?"}]
ai_msg = model_with_tools.invoke(message)
message.append(ai_msg)

#step 2 Exicute tools and collect results
for tool_call in ai_msg.tool_calls:
    # Execute the tool call
    tool_result = get_weather.invoke(tool_call)
    message.append(tool_result)
#step 3 pass result back to model for final response
final_response = model_with_tools.invoke(message)
print(final_response.text)
```
