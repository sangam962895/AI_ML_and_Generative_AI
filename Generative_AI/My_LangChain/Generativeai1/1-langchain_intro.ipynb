**What is Streaming and Batch**

For example, ChatGPT displays the response gradually as it is generated rather than showing the complete answer after a long wait. LangChain provides methods like stream() and astream() for this.

```python
model.invoke("write a paragraph for me")
```

```python
for chunk in model.stream("give 5 interviews question from the langchain"):
    print(chunk.text, end="", flush=True)
```

*Batch*

batch is a collection of indpendent request to a model can significantly improve performance and reduce costs, as the processing can be do in parrel

```python
responses = model.batch(
    [
        "tell me different b/w straming and batching"
        "explain streaming"
        "explain batching"
        "advantage of batching and straming"
        "beak the paragraph when one response output completed"
    ]
)
for response in responses:
    print(response.content)
```
