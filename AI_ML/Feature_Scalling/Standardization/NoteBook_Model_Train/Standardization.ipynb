# **StandardScale**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train_scaled # give numpay array
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
# ==========================================================
# STEP 1: Convert Scaled Arrays Back To DataFrame
# ==========================================================
# Problem:
# StandardScaler.transform() DataFrame return nahi karta.
#
# Ye NumPy Array return karta hai.
#
# Before:
# Age | Salary
#
# After:
# [[-1.2, 0.5],
#  [ 0.8,-0.3]]
#
# Column names gayab ho jate hain.
#
# Solution:
# Array ko wapas DataFrame me convert karo.
#
# Interview:
# Q. Why convert scaled data back to DataFrame?
# Ans:
# Better readability and column names preserve karne ke liye.

np.round(X_train.describe(), 1)
# STEP 2: Original Data Statistics Check
# ==========================================================
# describe()
#
# Summary statistics deta hai:
#
# count -> Total rows
# mean  -> Average
# std   -> Standard deviation
# min   -> Minimum value
# 25%   -> First Quartile
# 50%   -> Median
# 75%   -> Third Quartile
# max   -> Maximum value
#
# Why?
# Scaling se pehle data ka behaviour samajhne ke liye.
#
# Example:
#
# Age Range:
# 18 -> 60
#
# Salary Range:
# 15000 -> 150000
#
# Clearly Salary ki range bahut badi hai.
#
# Interview:
# Q. Why check statistics before scaling?
# Ans:
# Features ki scale aur distribution samajhne ke liye.
```

```python
np.round(X_train.describe(), 1)
# ==========================================================
# STEP 3: Check Scaled Data Statistics
# ==========================================================
# Goal:
# Verify karna ki StandardScaler ne sahi kaam kiya ya nahi.
#
# Expected Result:
#
# Mean ≈ 0
# Std  ≈ 1
#
# StandardScaler ka formula:
#
# z = (x - mean) / std
#
# Isliye scaling ke baad:
#
# Mean  -> 0
# Std   -> 1
#
# Important:
#
# Exact 0 aur 1 nahi bhi aa sakta.
#
# Example:
#
# Mean = 0.00000001
# Std  = 0.99999999
#
# Ye floating point precision issue hai.
#
# Interview:
# Q. How do you verify StandardScaler worked?
# Ans:
# Using describe() check:
#
# Mean ≈ 0
# Std ≈ 1
```

# **Effect of Scalling**

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")

ax2.scatter(X_train_scaled['Age'],
            X_train_scaled['EstimatedSalary'],
            color='red')
ax2.set_title("After Scaling")

plt.show()
# ==========================================================
# STEP 4: Visual Verification Using Scatter Plot
# ==========================================================
# Theory:
#
# Abhi tak humne mathematically verify kiya tha:
#
# Mean ≈ 0
# Std ≈ 1
#
# Lekin Data Science me sirf numbers dekhna enough nahi hota.
#
# Hume visually bhi dekhna chahiye ki scaling ke baad
# data me kya change aaya.
#
# Scatter Plot:
#
# Har dot = 1 user
#
# X-axis = Age
# Y-axis = EstimatedSalary
#
# Important:
#
# StandardScaler data ki shape change nahi karta.
#
# Example:
#
# Before:
#      *
#   *     *
#       *
#
# After:
#      *
#   *     *
#       *
#
# Pattern same rahega.
#
# Sirf scale change hoga.
#
# Why?
#
# StandardScaler relative distance preserve karta hai.
#
# Ye data ko stretch ya compress karta hai,
# lekin relationship nahi todta.
#
# Interview:
#
# Q. Does StandardScaler change data distribution?
#
# Ans:
# No.
#
# It changes scale only.
#
# Q. Does StandardScaler change shape?
#
# Ans:
# No.
#
# Only feature scale changes.
#
# Real Life Example:
#
# India ka map agar cm me measure karo
# ya inches me.
#
# Country ka shape same rahega.
#
# Sirf measurement unit change hogi.
#
# StandardScaler bhi wahi karta hai.
```

```python
# Scatter Plot = Visual verification

# Before Scaling:
# Different feature ranges

# After Scaling:
# Same feature ranges

# Shape = Same
# Scale = Changed

# StandardScaler:
# Changes scale
# Does NOT change pattern
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)
# ==========================================================
# INTERPRETATION OF KDE PLOT
# ==========================================================
#
# Left Graph:
#
# Original Data
#
# Age Distribution:
# Around 18 → 60
#
# Salary Distribution:
# Around 15000 → 150000
#
# Dono curves different scale par hain.
#
#
# Right Graph:
#
# After StandardScaler
#
# Age:
# Around -3 → +3
#
# Salary:
# Around -3 → +3
#
#
# Observe:
#
# Shape SAME hai.
#
# Curve SAME hai.
#
# Peak SAME hai.
#
#
# Sirf X-axis values change hui hain.
#
#
# This proves:
#
# StandardScaler
#
# ✔ Scale change karta hai
#
# ✘ Distribution shape nahi badalta
#
#
# Interview:
#
# Q. Why does KDE shape remain same?
#
# Ans:
# StandardScaler performs linear transformation.
#
# It does not alter the underlying distribution.
#
#
# Q. What changed?
#
# Ans:
# Feature scale
#
#
# Q. What did not change?
#
# Ans:
# Distribution shape
# Relative pattern
# Data ordering

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
# ==========================================================
# AGE DISTRIBUTION ANALYSIS
# ==========================================================
#
# Before Scaling:
#
# Age Range:
# 18 → 60
#
#
# After Scaling:
#
# Age Range:
# Approximately -3 → +3
#
#
# Observe:
#
# Bell Shape SAME
#
# Peak SAME
#
# Density SAME
#
#
# Only axis values changed.
#
#
# Why?
#
# StandardScaler:
#
# z = (x - mean) / std
#
# Har value ko relative scale par convert karta hai.
#
#
# Example:
#
# Age = 20
#
# Mean = 30
#
# Std = 10
#
#
# z-score:
#
# (20 - 30)/10
#
# = -1
#
#
# Meaning:
#
# This value is 1 standard deviation below the mean.
#
#
# Interview:
#
# Q. What is a z-score?
#
# Ans:
# Number of standard deviations a value is away from the mean.
#
#
# Q. Why is StandardScaler sometimes called Z-score Normalization?
#
# Ans:
# Because it uses:
#
# z = (x - mean)/std

plt.show()
```

# Why Scaling is **Important**

```python
from sklearn.linear_model import LogisticRegression
```

```python
# ==========================================================
# STEP 7: LOGISTIC REGRESSION
# ==========================================================
#
# Story:
#
# Ab tak humne data ko clean kiya.
# Train-Test Split kiya.
# StandardScaler lagaya.
# Ab question:
#
# Kya scaling se model ki performance improve hui?
#
# Isko check karne ke liye hum Logistic Regression use karenge.
# Logistic Regression:
#
# Classification Algorithm
#
# Use Cases:
#
# Email Spam or Not Spam
# Customer Buy Karega Ya Nahi
# Loan Approve Hoga Ya Nahi
# Disease Hai Ya Nahi
#
#
# Humare Case Me:
#
# Purchased
#
# 0 = Not Purchased
# 1 = Purchased
#
#
# Interview:
#
# Q. Logistic Regression Regression hai ya Classification?
#
# Ans:
# Classification Algorithm
#
#
# Q. Output kya deta hai?
#
# Ans:
# Probability
#
# Example:
#
# 0.95
#
# Means:
#
# 95% chance user product kharidega
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
# ==========================================================
# STEP 8: MODEL OBJECT CREATE KARNA
# ==========================================================
#
# Ab hum do models banayenge.
#
# Model 1:
# Original Data
#
# Model 2:
# Scaled Data
#
# Purpose:
#
# Compare karna ki scaling ka effect kya hua.
#
#
# Interview:
#
# Q. Why create two models?
#
# Ans:
# To compare performance before and after scaling.
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
# ==========================================================
# STEP 9: TRAINING THE MODEL
# ==========================================================
#
# fit()
#
# Most Important Function
#
# Meaning:
#
# Learn
#
#
# Machine Learning Flow:
#
# X_train
#
# Features
#
# Age
# Salary
#
# +
#
# y_train
#
# Purchased
#
#
# Model Pattern Learn Karta Hai
#
#
# Example:
#
# Age = High
# Salary = High
#
# Purchased = 1
#
#
# Age = Low
# Salary = Low
#
# Purchased = 0
#
#
# Model gradually relationship seekhta hai.
#
#
# Interview:
#
# Q. What happens inside fit()?
#
# Ans:
# Model learns parameters from training data.
#
#
# Q. Does fit() memorize data?
#
# Ans:
# No.
#
# It learns patterns.
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
# ==========================================================
# STEP 10: PREDICTION
# ==========================================================
#
# predict()
#
# Meaning:
#
# Future prediction
#
#
# Training Complete
#
# Ab exam shuru.
#
#
# X_test
#
# New unseen users
#
#
# Example:
#
# Age = 45
# Salary = 90000
#
#
# Model predict karega:
#
# 1
#
# Means:
#
# User purchase karega.
#
#
# Interview:
#
# Q. Difference between fit() and predict()?
#
# fit()
#
# Learn
#
#
# predict()
#
# Predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual", accuracy_score(y_test, y_pred))
print("Scaled", accuracy_score(y_test, y_pred_scaled))
# ==========================================================
# STEP 12: PERFORMANCE COMPARISON
# ==========================================================
#
# Purpose:
#
# Compare
#
# Original Data
#
# vs
#
# Scaled Data
#
#
# Possible Output:
#
# Actual  = 0.875
#
# Scaled  = 0.8666
#
#
# Important:
#
# Scaling ALWAYS accuracy improve nahi karti.
#
#
# Common Myth:
#
# Scaling = Higher Accuracy
#
# ❌ Wrong
#
#
# Reality:
#
# Scaling helps optimization.
#
# Accuracy increase ho bhi sakti hai.
#
# Nahi bhi ho sakti.
#
#
# Interview:
#
# Q. Does StandardScaler always improve accuracy?
#
# Ans:
# No.
#
# It improves feature scaling,
# not necessarily model performance.
```

```python
dt.fit(X_train, y_train)
dt_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print("Actual", accuracy_score(y_test, y_pred))
print("Scaled", accuracy_score(y_test, y_pred_scaled))
```
